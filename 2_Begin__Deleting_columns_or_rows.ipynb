### Load required packages and data

```
# Import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Save Github location paths to a variable
failed_bank_path = 'https://github.com/The-Calculated-Life/python_analysis_for_excel/blob/main/data/failed_banks.xlsx?raw=true'
bx_books_path = 'https://raw.githubusercontent.com/The-Calculated-Life/python_analysis_for_excel/main/data/bx_books.csv'

# Read excel and CSV files
bank_detail = pd.read_excel(failed_bank_path, sheet_name='detail')
bx_books = pd.read_csv(bx_books_path)
```

<br><br>
### Remove columns (axis=1)
---

```
# View first 5 rows of bank_detail
bank_detail.head()
```

<br>

```
# Remove the CHARTER column (without axis keyword)
bank_detail.drop(['CHARTER'])
```

<br>

```
# Correct the error
bank_detail.drop(['CHARTER'], axis=1)
```

<br>

```
# Look at the first 5 rows again
bank_detail.head()
```

<br>

```
# Save the changes
bank_detail = bank_detail.drop(['CHARTER'], axis=1)
```

<br>

```
# View the .head() again
bank_detail.head()
```

### Remove rows (axis=0)
---



```
# Look at dataframe
bank_detail.head()
```

<br>

```
# Remove the visable rows with missing "ESTIMATED LOSS" values
bank_detail.drop([0,1])
```

<br>

```
# Find rows with missing ESTIMATED LOSS
missing_loss = bank_detail['ESTIMATED LOSS'].isnull()
```

<br>

```
# Find index values for rows with missing ESTIMATED LOSS
bank_detail[missing_loss].index
```

<br>

```
# Save index values to variable
rows_to_drop = bank_detail[missing_loss].index
```

<br>

```
# Drop rows with missing ESTIMATED LOSS
bank_detail = bank_detail.drop(rows_to_drop)
```

<br>

```
# Confirm that the drop occured
bank_detail.head()
```
