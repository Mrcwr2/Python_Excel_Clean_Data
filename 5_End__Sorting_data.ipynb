# Sorting Data
---

## **Excel:**

*   Can sort using ribbon in ascending or descending order
*   Custom sort allows for sorting by multiple columns

sort-data.png

## **Python:**



*   Sort through use of .sort_values() function
*   Sort by ascending or descending (ascending is default)
*   Can sort by multiple conditions
*   Can sort by a column or index


    df.sort_values('column', ascending=False)

OR

    df.sort_index(ascending=False)


<br>

### Load required packages and data
---

```
# Import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Save Github location paths to a variable
failed_bank_path = 'https://github.com/The-Calculated-Life/python_analysis_for_excel/blob/main/data/failed_banks.xlsx?raw=true'
bx_ratings_path = 'https://raw.githubusercontent.com/The-Calculated-Life/python_analysis_for_excel/main/data/bx_ratings.csv'

# Read excel and CSV files
bank_detail = pd.read_excel(failed_bank_path, sheet_name='detail')
bank_list = pd.read_excel(failed_bank_path, sheet_name='banks')
bank_dividends = pd.read_excel(failed_bank_path, sheet_name='dividends')
bx_ratings = pd.read_csv(bx_ratings_path)
```

<br>

### Sorting numbers & index
---

```
# Sort "bank_detail" so that banks with the least ASSETS appear first
bank_detail.sort_values('ASSETS')
```

<br>

```
# Sort "bank_detail" so that banks with the most DEPOSITS appear first
bank_detail.sort_values('DEPOSITS', ascending=False)
```

<br>

```
# Save as dataframe "deposits_descending_df"
deposits_descending_df = bank_detail.sort_values('DEPOSITS', ascending=False)
```

<br>

```
# View the result
deposits_descending_df.head()
```

<br>

```
# Return the sorted dataframe to its original sorting based on index
deposits_descending_df.sort_index()
```

<br>

### Sorting strings & index
---

```
# Sort bank_list alphabetically by ST and City (ascending)
bank_list.sort_values(['ST', 'City'])
```

<br>

```
# Sort by ST ascending but City descending
bank_list.sort_values(['ST', 'City'], ascending=[True, False])
```

<br>

### Sort by date index
---

```
# Another example: Set the index and sort by it
bank_list.set_index('Closing Date', inplace=True)
```

<br>

```
# Set and sort index in one line
bank_list.sort_index()
```
