# Changing data types
---

## **Excel:**

*   Determined when you load or enter data
*   Can be changed on the ribbon

change-data-type.png


<br>

## **Python:**



*   Determined when you load data or create a variable
*   Changed through code


**One value:**

    bool(value)
    str(value)
    int(value)
    float(value)

    Check type with: type() function

**Series:**

    value.astype(bool)
    value.astype(str)
    value.astype(int)
    value.astype(float)

    Check type with: .dtypes attribute

**Dates**

    pd.to_datetime(value)

<br>

### Load required packages and data
---

```
# Import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Save Github location paths to a variable
failed_bank_path = 'https://github.com/The-Calculated-Life/python_analysis_for_excel/blob/main/data/failed_banks.xlsx?raw=true'
bx_ratings_path = 'https://raw.githubusercontent.com/The-Calculated-Life/python_analysis_for_excel/main/data/bx_ratings.csv'

# Read excel and CSV files
bank_detail = pd.read_excel(failed_bank_path, sheet_name='detail')
bank_list = pd.read_excel(failed_bank_path, sheet_name='banks')
bx_ratings = pd.read_csv(bx_ratings_path)
```

### Data types in dataframes
---

```
# View bank_list
bank_list.head()
```

<br>

```
# Check data type of a series
bank_list['CERT'].dtype
```

<br>

```
# Wrap series in str()
str(bank_list['CERT'])
```

<br>

```
# Use the appropriate method instead
bank_list['CERT'].astype(str)
```

<br>

```
# Save the change in data type
bank_list['CERT'] = bank_list['CERT'].astype(str)
```

<br>

```
# Check data type of first row in series
type(bank_list['CERT'][0])
```

<br>

### Working with dates
---

```
# Check the type of the "python_3_released" variable
python_3_released = '2008-12-03'
```

<br>

```
# Convert "python_3_released" to a date
pd.to_datetime(python_3_released)
```

<br>

```
# Convert "Closing Date" in bank_list
bank_list['Closing Date'] = pd.to_datetime(bank_list['Closing Date'])
```

<br>

```
# View results
bank_list.head()
```
